1.เขียนโปรแกรมสำหรับสร้างแบบจำลองเชิงเส้นด้วยวิธีลดตามความชัน พร้อมทั้งแสดงฟังก์ชันค่าใช้จ่ายในรูปของคอนทัวร์และแสดงให้เห็นถึงขั้นตอนในการปรับพารามิเตอร์

```python
def CalculateMSE(x_positions, y_positions, w0, w1):
    n = len(x_positions)
    MSE = 0
    for i in range(0, n):
        MSE += (w0 + w1*x_positions[i] - y_positions[i]) ** 2
    MSE *= 1/(2*n)
    return MSE

def GradientDescent(x_positions, y_positions, a, w0, w1):
    n = len(x_positions)
    w0_new = w0
    w1_new = w1
    for i in range(0,n): 
        w0_new -= (a / n) * (w0 + w1*x_positions[i] - y_positions[i])
        w1_new -= (a / n) * (w0 + w1*x_positions[i] - y_positions[i]) * x_positions[i]
    return (w0_new ,w1_new)
```

```python
df = px.data.tips()
x_input = df['total_bill']
y_input = df['tip']
#x = [47,72,21,19,72,56,49,89,43,66,41,33,23,32,80,31,78,21,43,45,61,62,18,44,30,65,78,71,60,58,71,33,59,81,71,75,64,41,82,60]
#y = [38,9,10,10,52,14,66,17,18,21,22,25,31,32,33,33,33,39,41,43,43,47,49,49,52,52,59,63,68,75,79,79,80,86,89,90,95,95,100,100] 
a = 0.001
w0 = 1
w1 = 0
w0_list = [w0]
w1_list = [w1]
E = [CalculateMSE(x_input, y_input, w0, w1)]
print(CalculateMSE(x_input, y_input, w0, w1))
for i in range(0,10) :
    w0, w1 = GradientDescent(x_input, y_input, a, w0, w1)
    w0_list.append(w0)
    w1_list.append(w1)
    #print(CalculateMSE(x, y, w0, w1))
    E.append(CalculateMSE(x_input, y_input, w0, w1))
    
print(w0, w1)
print(E)
```

```python
def Standardize(points):
    a = 0
    n = len(points)
    mean = sum(points) / n
    new_points = []
    for i in range(0,n):
        a += (points[i] - mean) ** 2
    std = math.sqrt(a / (n-1))
    for i in range(0,n):
        new_points.append((points[i] - mean) / std)
    return new_points
```
